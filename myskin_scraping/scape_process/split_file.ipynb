```python
# split df into 10 part and save to "/Users/naruebet.orgl/Workspace/Labs/rnd_ai/myskin_scraping/raws_split"
import numpy as np

# Calculate the size of each split
total_rows = len(df)
num_splits = 10
chunk_size = total_rows // num_splits

print(f"Total rows: {total_rows}")
print(f"Approximate rows per split: {chunk_size}")

# Split the dataframe into 10 parts
df_splits = np.array_split(df, num_splits)

# Save each split to a separate file
output_dir = "/Users/naruebet.orgl/Workspace/Labs/rnd_ai/myskin_scraping/raws_split"

for i, split_df in enumerate(df_splits, 1):
    output_file = f"{output_dir}/split_{i}_of_{num_splits}.json"
    split_df.to_json(output_file, orient='records', force_ascii=False, indent=2)
    print(f"Saved split {i}: {len(split_df)} rows -> {output_file}")

print(f"\nAll {num_splits} splits saved successfully!")
```

```python
import numpy as np
import os

# Calculate the size of each split
total_rows = len(df)
num_splits = 10
chunk_size = total_rows // num_splits

print(f"Total rows: {total_rows}")
print(f"Approximate rows per split: {chunk_size}")

# Split the dataframe into 10 parts
df_splits = np.array_split(df, num_splits)

# Output directory
output_dir = "/Users/naruebet.orgl/Workspace/Labs/rnd_ai/myskin_scraping/raws_split"
os.makedirs(output_dir, exist_ok=True)

# Save each split to a separate Excel file
for i, split_df in enumerate(df_splits, 1):
    output_file = f"{output_dir}/split_{i}_of_{num_splits}.xlsx"
    split_df.to_excel(output_file, index=False)
    print(f"Saved split {i}: {len(split_df)} rows -> {output_file}")

print(f"\nAll {num_splits} Excel splits saved successfully!")
```
