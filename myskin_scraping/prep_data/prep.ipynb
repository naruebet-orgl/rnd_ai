```python
# 🔍 Dynamic Column Subset Cross-Checker for Large DataFrames
# ----------------------------------------------------------
# Randomly sample 200–300 rows and test whether each column
# is a subset of every other column in the same DataFrame.
# Works on big scraped datasets with noisy text.

import re
import math
import unicodedata
from typing import List, Optional, Tuple
import numpy as np
import pandas as pd
from difflib import SequenceMatcher
from IPython.display import display

# ---------- Normalization & Tokenization ----------
def normalize_text(s: Optional[str]) -> str:
    if s is None or (isinstance(s, float) and math.isnan(s)):
        return ""
    s = unicodedata.normalize("NFKC", str(s))
    s = s.lower().strip()
    s = re.sub(r"\s+", " ", s)
    return s

def word_tokens(s: str) -> List[str]:
    return re.findall(r"[a-zA-Z0-9ก-๙]+", s.lower())

# ---------- Similarity Functions ----------
def substring_contains(small: str, big: str) -> bool:
    return small in big if small and big else False

def token_coverage(small: str, big: str) -> float:
    t_small = set(word_tokens(small))
    if not t_small:
        return 1.0
    t_big = set(word_tokens(big))
    return len(t_small & t_big) / len(t_small)

def jaccard_tokens(a: str, b: str) -> float:
    ta, tb = set(word_tokens(a)), set(word_tokens(b))
    if not ta and not tb:
        return 1.0
    return len(ta & tb) / len(ta | tb) if (ta | tb) else 0.0

def fuzzy_ratio(a: str, b: str) -> float:
    if not a and not b:
        return 1.0
    return SequenceMatcher(None, a, b).ratio()

# ---------- Sampling & Filtering ----------
def _prepare_sample(df: pd.DataFrame, sample_n: int, random_state: int) -> pd.DataFrame:
    n = min(sample_n, len(df))
    if n <= 0:
        return df.head(0).copy()
    return df.sample(n=n, random_state=random_state)

def _textlike_series(s: pd.Series) -> bool:
    if pd.api.types.is_string_dtype(s) or s.dtype == object:
        return True
    non_null = s.dropna().astype(str)
    if non_null.empty:
        return False
    letter_frac = np.mean(non_null.str.contains(r"[A-Za-zก-๙]", regex=True, na=False))
    return letter_frac > 0.6

# ---------- Evaluation Logic ----------
def _evaluate_pair(sample: pd.DataFrame, cand: str, cont: str,
                   token_cov_threshold: float, fuzzy_threshold: float) -> Tuple[float, float, float, float, int]:
    c_small = sample[cand].map(normalize_text)
    c_big = sample[cont].map(normalize_text)
    sub_ok = [substring_contains(a, b) for a, b in zip(c_small, c_big)]
    tok_cov = [token_coverage(a, b) for a, b in zip(c_small, c_big)]
    fuzz = [fuzzy_ratio(a, b) for a, b in zip(c_small, c_big)]
    decision = np.array(sub_ok) | ((np.array(tok_cov) >= token_cov_threshold) & (np.array(fuzz) >= fuzzy_threshold))
    return (float(np.mean(sub_ok)),
            float(np.mean(tok_cov)),
            float(np.mean(fuzz)),
            float(np.mean(decision)),
            len(sample))

# ---------- Main Cross-Check ----------
def crosscheck_all_subsets(
    df: pd.DataFrame,
    sample_n: int = 300,
    random_state: int = 42,
    token_cov_threshold: float = 0.8,
    fuzzy_threshold: float = 0.6,
    exclude_numeric: bool = True,
) -> pd.DataFrame:
    # select only text-like columns if exclude_numeric=True
    cols = [c for c in df.columns if not exclude_numeric or _textlike_series(df[c])]
    sample = _prepare_sample(df[cols], sample_n=sample_n, random_state=random_state)

    results = []
    for cand in cols:
        for cont in cols:
            if cand == cont:
                continue
            try:
                sub_hr, avg_tok, avg_fz, pass_rt, nrows = _evaluate_pair(
                    sample[[cand, cont]].dropna(how="all"),
                    cand=cand, cont=cont,
                    token_cov_threshold=token_cov_threshold,
                    fuzzy_threshold=fuzzy_threshold
                )
                results.append({
                    "candidate": cand,
                    "container": cont,
                    "substring_hit_rate": round(sub_hr, 3),
                    "avg_token_coverage": round(avg_tok, 3),
                    "avg_fuzzy_ratio": round(avg_fz, 3),
                    "pass_rate": round(pass_rt, 3),
                    "rows_evaluated": nrows
                })
            except Exception as e:
                results.append({
                    "candidate": cand, "container": cont,
                    "error": str(e), "rows_evaluated": 0
                })

    summary = pd.DataFrame(results).sort_values(
        ["pass_rate", "substring_hit_rate", "avg_token_coverage"],
        ascending=False
    ).reset_index(drop=True)
    display(summary)
    return summary

# ---------- Drilldown: Inspect Mismatches ----------
def inspect_mismatches(
    df: pd.DataFrame,
    candidate_col: str,
    container_col: str,
    sample_n: int = 300,
    random_state: int = 42,
    token_cov_threshold: float = 0.8,
    fuzzy_threshold: float = 0.6,
) -> pd.DataFrame:
    sample = _prepare_sample(df[[candidate_col, container_col]], sample_n, random_state)
    sample["__cand__"] = sample[candidate_col].map(normalize_text)
    sample["__cont__"] = sample[container_col].map(normalize_text)

    sub_ok = sample.apply(lambda r: substring_contains(r["__cand__"], r["__cont__"]), axis=1)
    tok_cov = sample.apply(lambda r: token_coverage(r["__cand__"], r["__cont__"]), axis=1)
    fuzz = sample.apply(lambda r: fuzzy_ratio(r["__cand__"], r["__cont__"]), axis=1)
    decision = sub_ok | ((tok_cov >= token_cov_threshold) & (fuzz >= fuzzy_threshold))

    mism = sample.loc[~decision, [candidate_col, container_col]].copy()
    mism["__substring__"] = sub_ok[~decision].values
    mism["__token_cov__"] = np.round(tok_cov[~decision].values, 3)
    mism["__fuzzy__"] = np.round(fuzz[~decision].values, 3)
    display(mism)
    return mism

# ---------- Example Usage ----------
# summary = crosscheck_all_subsets(df, sample_n=300)
# inspect_mismatches(df, candidate_col='a', container_col='y', sample_n=300)
```

```python
import re
import math
import numpy as np
import pandas as pd

# ---------- A) กรองแถว noise: "ตะกร้า" + URL listing ----------
_LISTING_HINTS = [
    r"[?&]page=\d+",
    r"[?&]order=",
    r"/category",  # กันเผื่อ
    r"/search",
]
def is_listing_row(name: str, url: str, category: str) -> bool:
    name = (name or "").strip()
    url = (url or "").strip().lower()
    category = (category or "").strip().lower()
    if name in ("ตะกร้า", "Cart", "Shopping Cart"):
        return True
    for pat in _LISTING_HINTS:
        if re.search(pat, url):
            return True
    # บางเว็บยัด query ลง column category ตรง ๆ
    if re.search(r"[?&](page|order)=", category):
        return True
    return False

# ---------- B) แปลง "None" เป็น NaN ----------
def none_string_to_nan(df: pd.DataFrame) -> pd.DataFrame:
    return df.replace(to_replace=r"^\s*None\s*$", value=np.nan, regex=True)

# ---------- C) (ย้ำ) ตัวแยกราคาเวอร์ชันก่อนหน้า ----------
_CURRENCY_PAT = r"(?:฿|THB|USD|\$|EUR|€)"
_UNIT_KEYWORDS = [
    (r"(?:per\s*kg|ต่อ\s*kg|/\s*kg|\bkg\b|กก\.?)", "per_kg"),
    (r"(?:per\s*g|ต่อ\s*กรัม|/\s*g|\bg\b|กรัม)", "per_g"),
    (r"(?:per\s*ml|ต่อ\s*ml|/\s*ml|\bml\b|มล\.?|มิลลิลิตร)", "per_ml"),
    (r"(?:per\s*l|ต่อ\s*l|/\s*l|\bl\b|ลิตร)", "per_l"),
    (r"(?:ต่อ\s*ชิ้น|per\s*piece|/\s*pc|\bpcs?\b|ชิ้น)", "per_piece"),
    (r"(?:ต่อ\s*ขวด|bottle|ขวด)", "per_bottle"),
    (r"(?:ต่อ\s*แพ็ค|per\s*pack|แพ็ค|แพค|pack)", "per_pack"),
    (r"(?:ต่อ\s*หน่วย|per\s*unit|ต่อ\s*ยูนิต|unit)", "per_unit"),
]
def _norm_number(s: str) -> float:
    s = s.strip()
    if "," in s and "." in s:
        if s.rfind(".") > s.rfind(","):
            s = s.replace(",", "")
        else:
            s = s.replace(".", "").replace(",", ".")
    else:
        s = s.replace(",", "")
    try:
        return float(s)
    except:
        return math.nan

def _find_all_prices(text: str):
    out = []
    if not isinstance(text, str) or not text:
        return out
    pat_curr = re.compile(rf"(?P<cur>{_CURRENCY_PAT})\s*(?P<num>[\d.,]+)", re.I)
    for m in pat_curr.finditer(text):
        out.append((m.start(), m.end(), m.group("cur"), m.group("num"), _norm_number(m.group("num")), m.group(0)))
    pat_num = re.compile(r"(?<![A-Za-z0-9])(?P<num>[\d]{1,3}(?:[\d,\.]*\d)?)(?![A-Za-z%])")
    for m in pat_num.finditer(text):
        snippet = text[max(0, m.start()-5):m.end()+5]
        if "%" in snippet: 
            continue
        val = _norm_number(m.group("num"))
        if not math.isnan(val):
            out.append((m.start(), m.end(), "", m.group("num"), val, m.group(0)))
    return out

def _detect_unit_near(text: str, idx: int, window: int = 30) -> str:
    start = max(0, idx - window); end = min(len(text), idx + window)
    ctx = text[start:end]
    for pat, unit in _UNIT_KEYWORDS:
        if re.search(pat, ctx, flags=re.I):
            return unit
    if re.search(r"(?:ราคา\s*ต่อ|price\s*per|per\s*/?)", ctx, flags=re.I):
        return "per_unit"
    return ""

def _score_price_candidate(text: str, cand) -> tuple:
    i, j, cur, num_str, val, matched = cand
    score = 0
    if cur: score += 2
    unit = _detect_unit_near(text, i, window=40)
    if unit: score += 1
    vicinity = text[max(0, i-20):min(len(text), j+20)]
    if re.search(r"(ราคา|price)", vicinity, flags=re.I): score += 1
    return score, unit

def extract_price_general(text: str):
    res = {
        "price_value": np.nan, "price_min": np.nan, "price_max": np.nan, "is_range": False,
        "currency": "", "unit": "", "matched_text": "", "notes": "",
    }
    if not isinstance(text, str) or not text.strip():
        return res
    cands = _find_all_prices(text)
    if not cands:
        return res
    range_pat = re.compile(
        rf"(?P<cur1>{_CURRENCY_PAT})?\s*(?P<a>[\d.,]+)\s*[-–~]\s*(?P<cur2>{_CURRENCY_PAT})?\s*(?P<b>[\d.,]+)",
        re.I
    )
    m_range = range_pat.search(text)
    if m_range:
        a = _norm_number(m_range.group("a")); b = _norm_number(m_range.group("b"))
        cur = m_range.group("cur1") or m_range.group("cur2") or ""
        unit = _detect_unit_near(text, m_range.start())
        res.update({
            "price_min": a, "price_max": b, "is_range": True,
            "currency": cur, "unit": unit, "matched_text": m_range.group(0),
            "notes": "parsed_range"
        })
        return res
    scored = []
    for cand in cands:
        score, unit = _score_price_candidate(text, cand)
        scored.append((score, unit, cand))
    scored.sort(key=lambda x: x[0], reverse=True)
    best_score, unit, best = scored[0]
    i, j, cur, num_str, val, matched = best
    res.update({
        "price_value": val, "currency": cur or "THB", "unit": unit or "",
        "matched_text": matched, "notes": f"score={best_score}; total_candidates={len(cands)}"
    })
    return res

def apply_price_parsing(df: pd.DataFrame, price_col="price"):
    out = df[price_col].apply(extract_price_general).apply(pd.Series)
    return pd.concat([df, out], axis=1)

# ---------- D1) parser สำหรับ usage range ----------
_USAGE_RANGE = re.compile(
    r"(?P<min>[\d.,]+)\s*[-–]\s*(?P<max>[\d.,]+)\s*%\b", re.I
)
def parse_usage_range(s: str):
    if not isinstance(s, str): 
        return pd.Series({"usage_min_pct": np.nan, "usage_max_pct": np.nan, "usage_matched": ""})
    m = _USAGE_RANGE.search(s)
    if not m:
        # single value % เช่น 1.0%
        m2 = re.search(r"(?P<val>[\d.,]+)\s*%\b", s)
        if m2:
            v = _norm_number(m2.group("val"))
            return pd.Series({"usage_min_pct": v, "usage_max_pct": v, "usage_matched": m2.group(0)})
        return pd.Series({"usage_min_pct": np.nan, "usage_max_pct": np.nan, "usage_matched": ""})
    mn = _norm_number(m.group("min")); mx = _norm_number(m.group("max"))
    return pd.Series({"usage_min_pct": mn, "usage_max_pct": mx, "usage_matched": m.group(0)})

# ---------- D2) ทำความสะอาดข้อความยาวซ้ำ ----------
def clean_redundant_text(s: str, max_repeat_ratio: float = 0.8):
    """
    ตัดย่อหน้าที่ซ้ำเกินไป: ถ้าประโยคเหมือนกันซ้ำ >80% ให้เก็บไว้ครั้งเดียว
    """
    if not isinstance(s, str) or not s.strip():
        return s
    # normalize space
    s_norm = re.sub(r"[ \t]+", " ", s)
    s_norm = re.sub(r"\n{3,}", "\n\n", s_norm)
    # de-duplicate sentences
    parts = [p.strip() for p in re.split(r"(?<=[\.\!\?]|[\u0E00-\u0E7F]\))\s+|\n+", s_norm) if p.strip()]
    seen = []
    out = []
    for p in parts:
        if any(len(os) >= int(len(p)*max_repeat_ratio) and os in p or p in os for os in seen):
            continue
        seen.append(p)
        out.append(p)
    return "\n".join(out)

def apply_text_cleaning(df: pd.DataFrame, cols=("description","benefits","detailed_info","usage")):
    for c in cols:
        if c in df.columns:
            df[c] = df[c].apply(clean_redundant_text)
    return df

# ---------- Pipeline Helper ----------
def preprocess_dataset(df: pd.DataFrame) -> pd.DataFrame:
    df = none_string_to_nan(df.copy())
    # กรอง listing rows
    mask_noise = df.apply(lambda r: is_listing_row(r.get("name"), r.get("url"), r.get("category")), axis=1)
    df = df.loc[~mask_noise].reset_index(drop=True)
    # ราคา
    df = apply_price_parsing(df, price_col="price")
    # usage %
    if "usage" in df.columns:
        usage_df = df["usage"].apply(parse_usage_range)
        df = pd.concat([df, usage_df], axis=1)
    # ทำความสะอาด text ยาว
    df = apply_text_cleaning(df)
    return df

# ---------- วิธีใช้ ----------
# df_clean = preprocess_dataset(df2)
# df_clean.loc[:, ["name","price","price_value","currency","unit","matched_text","usage","usage_min_pct","usage_max_pct"]].head(20)
```

```python
import re
from difflib import SequenceMatcher

def clean_redundant_text(s: str, sim_threshold: float = 0.80):
    """
    - ไม่ใช้ look-behind -> ปลอด error บน Python re
    - ตัดความซ้ำของข้อความยาว: ถ้าประโยค/ท่อน ข้อความคล้ายกัน >= sim_threshold จะเก็บไว้ครั้งเดียว
    - ทำความสะอาดช่องว่าง/บรรทัดว่างยาว ๆ
    """
    if not isinstance(s, str) or not s.strip():
        return s

    # 1) normalize space/newlines
    s_norm = re.sub(r"[ \t]+", " ", s)
    s_norm = re.sub(r"\n{3,}", "\n\n", s_norm)

    # 2) แบ่งเป็นท่อนแบบหยาบ (ไม่ใช้ look-behind)
    #    แยกด้วย .,!,?, จุดจบประโยคทั่วไป, และขึ้นบรรทัดใหม่
    #    (ภาษาไทยไม่เว้นวรรคจบประโยคชัดเจน ใช้ newline/จุด bullet ช่วย)
    parts = re.split(r"(?:[.!?]+|\n+|•|·|–|—|-{2,}|。|！|？|…)", s_norm)
    parts = [p.strip() for p in parts if p and p.strip()]

    # 3) de-duplicate ด้วย similarity + contain check
    out = []
    for p in parts:
        is_dup = any(
            SequenceMatcher(None, p, q).ratio() >= sim_threshold or (p in q) or (q in p)
            for q in out
        )
        if not is_dup:
            out.append(p)

    return "\n".join(out)
```

```python
# --- Postprocessing Cleanup Block ---

def finalize_dataset(df: pd.DataFrame) -> pd.DataFrame:
    """
    Final cleanup after preprocess_dataset():
    - Drop intermediate columns: price_min, price_max, is_range, matched_text, notes, etc.
    - Clean inci_name prefix (remove "Name:" / "INCI Name:")
    - Trim whitespace and normalize casing
    """
    df = df.copy()

    # 1. Drop intermediate columns if present
    drop_cols = ["price_min", "price_max", "is_range", "matched_text", "notes"]
    df = df.drop(columns=[c for c in drop_cols if c in df.columns], errors="ignore")

    # 2. Clean INCI Name field
    if "inci_name" in df.columns:
        df["inci_name"] = (
            df["inci_name"]
            .astype(str)
            .replace(r"^(?:INCI\s*Name:|Name:)\s*", "", regex=True)  # remove prefixes
            .replace(r"\s+", " ", regex=True)                       # collapse spaces
            .str.strip()
            .replace("None", np.nan)
        )

    # 3. (Optional) normalize category casing
    if "category" in df.columns:
        df["category"] = df["category"].astype(str).str.strip().str.lower()

    return df
```
